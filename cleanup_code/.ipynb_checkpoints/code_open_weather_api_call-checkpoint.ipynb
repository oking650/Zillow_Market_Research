```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from scipy.stats import linregress
import json
import requests

with open('climate_city_all (1).json') as f:
    data = json.load(f)

df = pd.json_normalize(data, record_path=None, meta=['id', 'name', 'state', 'country', ['coord', 'lon'], ['coord', 'lat']])
df.columns = ['ID', 'Name', 'State', 'Country', 'Longitude', 'Latitude']
df
```

```python
filtered_df = df[df['Country'] == 'US']
filtered_df['full_name'] = filtered_df['Name'] + ', ' + filtered_df['State']
filtered_df.reset_index(drop=True, inplace=True)
filtered_df
```

# Import CSV's

```python
days_pending_csv = 'Resources_Clean/clean_median_days_pending_week_all.csv'
days_to_pending_df = pd.read_csv(days_pending_csv)
```

```python
days_to_pending_df
```

```python
pending_cities_rough = pd.merge(days_to_pending_df, filtered_df, left_on=['Region Name'], right_on = ['full_name'])
pending_cities_rough
```

```python
pending_cities_reduced = pending_cities_rough[(pending_cities_rough['Size Rank'] < 10) & (pending_cities_rough['Date'] < '2023-01') & (pending_cities_rough['Date'] > '2019-12')]
```

```python
pending_cities_reduced.reset_index(drop=True, inplace=True)
pending_cities_reduced
```

```python
api_key = "048bfbdf25a77253843c286cd380e065"

# Function to make API calls and extract temperature information
def get_temperature(row):
    lat = row['Latitude']
    lon = row['Longitude']
    date = row['Date']

    api_url = f"https://api.openweathermap.org/data/3.0/onecall/day_summary?lat={lat}&lon={lon}&date={date}&appid={api_key}"

    try:
        response = requests.get(api_url)
        data = response.json()

        morning_temp = data['temperature']['morning']
        afternoon_temp = data['temperature']['afternoon']
        evening_temp = data['temperature']['evening']
        night_temp = data['temperature']['night']

        # Calculate the average temperature
        avg_temp = (morning_temp + afternoon_temp + evening_temp + night_temp) / 4
        return avg_temp
    except Exception as e:
        print(f"Error fetching data for {lat}, {lon}, {date}: {e}")
        return None
    
    
pending_cities_reduced['Temperature'] = pending_cities_reduced.apply(get_temperature, axis=1)

pending_cities_reduced.head()
```

```python
pending_cities_reduced['Temperature (Farenheit)'] = round(9 / 5 * (pending_cities_reduced['Temperature'] - 273.15) + 32, 3)
pending_cities_reduced.head()
```

```python
csv_file_path = "Resources_Clean/clean_temperature_cities.csv"

pending_cities_reduced.to_csv(csv_file_path)
```
