```python
# Dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import scipy as py
from scipy.stats import linregress
```

```python
days_to_close_csv = 'Resources_Clean/clean_median_days_to_close_week_all.csv'
days_pending_csv = 'Resources_Clean/clean_median_days_pending_week_all.csv'
sale_list_csv = 'Resources_Clean/clean_median_sale_to_list_week_all.csv'
median_list_price_csv = 'Resources_Clean/clean_median_list_price_week_all.csv'


days_to_closed_df = pd.read_csv(days_to_close_csv)
days_to_pending_df = pd.read_csv(days_pending_csv)
sale_list_df = pd.read_csv(sale_list_csv)
median_list_price_df = pd.read_csv(median_list_price_csv)
```

```python
merged_df = pd.merge(days_to_closed_df, days_to_pending_df, on=['Region Name', 'Date'])
merged_df = pd.merge(merged_df, sale_list_df, on=['Region Name', 'Date'])
merged_df = pd.merge(merged_df, median_list_price_df, on=['Region Name', 'Date'])
print(merged_df.columns)
```

```python
merged_reduced = merged_df[['Region ID_x',
                                       'Size Rank_x',
                                       'Region Name',
                                       'Region Type_x',
                                       'State Name_x',
                                       'Date', 
                                       'Median Days to Close',
                                       'Median Days to Close Rolling Avg',
                                       'Median Days Pending',
                                       'Median Days Pending Rolling Avg',
                                       'Median Sale to List Ratio',
                                       'Median Sale to List Ratio Rolling Avg']]

merged_reduced = merged_reduced.rename(columns=
                                       {'Region ID_x': 'Region ID',
                                        'Size Rank_x': 'Size Rank',
                                        'Region Type_x': 'Region Type',
                                        'State Name_x' : 'State Name'})


merged_reduced.head()
```

```python
column_data_types = merged_reduced.dtypes
merged_reduced['Date']=pd.to_datetime(merged_reduced['Date'])

print(column_data_types)
```

```python
merged_topsize = merged_reduced.loc[merged_reduced['Size Rank'] < 10]
```

```python
merged_topsize.pivot(index='Date', columns='Region Name', values='Median Days to Close').plot(kind='line')
plt.title('Median Days to Close Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Median Days to Close')
plt.legend(title='Region Name', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.show()
```

```python
merged_topsize.pivot(index='Date', columns='Region Name', values='Median Days Pending Rolling Avg').plot(kind='line')
plt.title('Median Days Pending Rolling Avg Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Median Days Pending Rolling Avg')
plt.legend(title='Region Name', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.show()
```

```python
merged_topsize.pivot(index='Date', columns='Region Name', values='Median Sale to List Ratio' ).plot(kind='line')
plt.title('Median Sale to List Ratio Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Median Sale to List Ratio')
plt.legend(title='Region Name', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.show()
```

```python
correlation = merged_reduced['Median Sale to List Ratio'].corr(merged_reduced['Median Days to Close'])

print(f"Correlation between Median Sale to List Ratio and Median Days to Close: {correlation}")
```

```python
x = merged_reduced['Median Days Pending']
y = merged_reduced['Median Sale to List Ratio']

slope, intercept, r_value, p_value, std_err = linregress(x, y)

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")
print(f"Standard Error: {std_err}")
```

```python
plt.scatter(merged_topsize['Median Days Pending'], merged_topsize['Median Sale to List Ratio'], alpha=0.5)
plt.title('Scatter Plot: Median Days to Close vs. Median Sale to List Ratio')
plt.xlabel('Median Days Pending')
plt.ylabel('Median Sale to List Ratio')
plt.show()
```
