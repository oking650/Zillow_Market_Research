```python
# Dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.stats import pearsonr
import numpy as np
```

```python
days_to_close_csv = 'Resources_Clean/clean_median_days_to_close_week_all.csv'
days_pending_csv = 'Resources_Clean/clean_median_days_pending_week_all.csv'
sale_list_csv = 'Resources_Clean/clean_median_sale_to_list_week_all.csv'

sale_price_csv = 'Resources_Clean/clean_median_sale_price_week_all.csv'

days_to_close_df = pd.read_csv(days_to_close_csv)
days_to_pending_df = pd.read_csv(days_pending_csv)
sale_list_df = pd.read_csv(sale_list_csv)
sale_price_df = pd.read_csv(sale_price_csv)
```

```python
merged_df = pd.merge(days_to_close_df, days_to_pending_df, on=['Region Name', 'Date'])
merged_df = pd.merge(merged_df, sale_list_df, on=['Region Name', 'Date'])
merged_df
```

```python
pending_salelist_df = pd.merge(sale_list_df, days_to_pending_df, on=['Region Name', 'Date'])
pending_salelist_df
```

```python
merged_reduced = pending_salelist_df[['Region ID_x',
                             'Size Rank_x',
                             'Region Name',
                             'Region Type_x',
                             'State Name_x',
                             'Date', 
                             'Median Days Pending',
                             'Median Days Pending Rolling Avg',
                             'Median Sale to List Ratio',
                             'Median Sale to List Ratio Rolling Avg']]

merged_reduced = merged_reduced.rename(columns = {'Region ID_x':'Region Id',
                             'Size Rank_x':'Size Rank',
                             'Region Type_x':'Region Type',
                             'State Name_x':'State Name',
                             'Median Days Pending':'Median Days to Pending',
                             'Median Days Pending Rolling Avg':'Median Days to Pending Rolling Avg'})

merged_reduced
```

```python
merged_reduced = merged_df[['Region ID_x',
                             'Size Rank_x',
                             'Region Name',
                             'Region Type_x',
                             'State Name_x',
                             'Date', 
                             'Median Days to Close',
                             'Median Days to Close Rolling Avg',
                             'Median Days Pending',
                             'Median Days Pending Rolling Avg',
                             'Median Sale to List Ratio',
                             'Median Sale to List Ratio Rolling Avg']]

merged_reduced = merged_reduced.rename(columns = {'Region ID_x':'Region Id',
                             'Size Rank_x':'Size Rank',
                             'Region Type_x':'Region Type',
                             'State Name_x':'State Name',
                             'Median Days Pending':'Median Days to Pending',
                             'Median Days Pending Rolling Avg':'Median Days to Pending Rolling Avg'})

merged_reduced
```

```python
merged_top_size = merged_reduced.loc[merged_reduced['Size Rank'] < 10]
```

```python
merged_top_size.pivot(index='Date', columns='Region Name', values='Median Days to Pending').plot(kind='line')
merged_top_size.pivot
plt.title('Median Days to Pending Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Median Days to Pending')
plt.legend(title='Region Name', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.show()
```

```python
merged_top_size.pivot(index='Date', columns='Region Name', values='Median Sale to List Ratio').plot(kind='line')
plt.title('Median Sale to List Ratio Over Time by Region')
plt.xlabel('Date')
plt.ylabel('Sale to List Ratio')
plt.legend(title='Region Name', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.show()
```

```python
correlation = merged_reduced['Median Sale to List Ratio'].corr(merged_reduced['Median Days to Pending'])

print(f"Correlation: {correlation}")
```

```python
# Selecting columns for linear regression
x = merged_reduced['Median Days to Pending']
y = merged_reduced['Median Sale to List Ratio']

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Print the regression results
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")
print(f"Standard Error: {std_err}")
```

```python
# Selecting columns for linear regression
x = merged_reduced['Median Days to Close']
y = merged_reduced['Median Sale to List Ratio']

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Print the regression results
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")
print(f"Standard Error: {std_err}")
```

```python
plt.scatter(merged_reduced['Median Days to Pending'], merged_reduced['Median Sale to List Ratio'], alpha=0.5)
plt.title('Scatter Plot: Median Days to Pending vs. Median Sale to List Ratio')
plt.xlabel('Median Days to Pending')
plt.ylabel('Median Sale to List Ratio')

z = np.polyfit(merged_reduced['Median Days to Pending'], merged_reduced['Median Sale to List Ratio'], 1)
p = np.poly1d(z)

plt.plot(merged_reduced['Median Days to Pending'], p(merged_reduced['Median Days to Pending']), color='red')

corr_coefficient, _ = pearsonr(merged_reduced['Median Days to Pending'], merged_reduced['Median Sale to List Ratio'])
plt.annotate(f'Correlation Coefficient: {corr_coefficient:.2f}', 
             xy=(0.8, 0.2), xycoords='axes fraction',
             fontsize=10, color='Red')
plt.show()
```
